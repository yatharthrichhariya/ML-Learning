```python
import numpy as np 
import pandas as pd 
```

```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder
```

```python
df=pd.read_csv('covid_19.csv')
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(df.drop(columns=['Tests']),df['Tests'],test_size=0.2)
```

```python
X_train
```

```python
from sklearn.compose import ColumnTransformer
```

```python
transformer = ColumnTransformer(transformers=[
    ('tnf1',SimpleImputer(),['Recovered']),
    ('tnf2',OrdinalEncoder(),['Cases']),
    ('tnf3',OneHotEncoder(sparse_output=False,drop='first'),['continent'])
],remainder='passthrough')
```

```python
transformer.fit_transform(X_train).shape
```

```python
transformer.fit_transform(X_test).shape
```
