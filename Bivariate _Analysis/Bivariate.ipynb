```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df=pd.read_csv('Titanic.csv')
```

```python
df
```

**Barplot**

```python
sns.barplot(y=df['Age'],x=df['Pclass'],palette='viridis')
```

```python
sns.barplot(y=df['Age'],x=df['Pclass'],hue=df['Sex'])
```

**Scatterplot**

```python
tips=pd.read_csv('tips.csv')
```

```python
sns.scatterplot(x=tips['total_bill'],y=tips['tip'],hue=tips['sex'],style=tips['smoker'],size=tips['size'])
```

**Boxplot**

```python
sns.boxplot(x=df['Sex'],y=df['Age'],hue=df['Survived'])
```

**Distplot**

```python
sns.distplot(df[df['Survived']==0]['Age'])
```

```python
sns.distplot(df[df['Survived']==0]['Age'],hist=False)
sns.distplot(df[df['Survived']==1]['Age'],hist=False)
```

**Heatmap**

```python
sns.heatmap(pd.crosstab(df["Pclass"],df["Survived"]))
```

```python
(df.groupby('Pclass')['Survived'].mean()*100).plot(kind='bar')
```

**Clustermap**

```python
sns.clustermap(pd.crosstab(df['SibSp'],df['Survived']))
```
